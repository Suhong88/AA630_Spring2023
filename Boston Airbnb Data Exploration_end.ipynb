# Boston Airbnb Open Data

This data set was obtained from [Kaggle](https://www.kaggle.com/datasets/airbnb/boston).

this dataset describes the listing activity of homestays in Boston, MA.

The following Airbnb activity is included in this Boston dataset:

 - Listings, including full descriptions and average review score
 - Reviews, including unique id for each reviewer and detailed comments
 - Calendar, including listing id and the price and availability for that day

# Adding/Removing columns

```python
# add a price indicator to show whether price is over 1000

df1['price_indicator']=df1.price_cleaned>1000
```

```python
df1[df1['price_indicator']]
```

```python
# create a column price_level, if price is less than 200, price level is Low, if price is between 200 and 500, the levei is medium, 
# if price is between 501 and 1000, the level is high, and if price is over 1000, the price is very high

def classify_price(price):
    if price<200:
        price_level="Low"
    elif price<500:
        price_level="Medium"
    elif price<1000:
        price_level="high"
    else:
        price_level="very high"
    return price_level

df1['price_level']=df1['price_cleaned'].apply(classify_price)
```
